### Analisa Deskriptif & Handling Missing Value / Outliers - 5 point

```python
df.isnull().sum() # mencari berapa banyak missing value (nilai adalah NaN) dari setiap kategori pada df
```

### Untuk memutuskan kategori mana yang akan Dropna atau Fillna, dicek dulu seberapa banyak data yang bernilai kosong (NaN atau NULL) dan apakah persentasenya signifikan terhadap data yang akan dianalisis. 

```python
df['category'].count() # Untuk mencari berapa total data yang tidak bernilai kosong (kategori produk yang dijual tidak memiliki missing value, diketahui dari hasil input df.isnull().sum())
```

### Dari hasil count, diketahui ada 611 data yang tidak bernilai kosong. Jika persentase missing value tidak signifikan (10 persen ke bawah), maka kategori tersebut dapat didrop (Dropna). Namun, ternyata kategori "perc_of_unit_sold" dan "rating" memiliki persentase missing value yang cukup signifikan, yaitu 14.57 % dan 30.12 %. Metode fillna dapat digunakan, tetapi akan mengandung error karena ada beberapa missing value yang di-fillna dengan mean sebagai nilai Fill-nya.  

### Shipping_rate dan gender tidak bernilai angka sehingga sangat sulit untuk menggunakan metode Fillna untuk kedua kategori tersebut. Dalam analisis E-Commerce ini, kedua variabel tersebut diabaikan. 

```python
dfdrop1 = df['Original_price'].dropna() # mencari nilai df['Original_price'] yang di-dropna
dfdrop1
```

```python
dfdrop1.describe() # melihat apakah variabel dfdrop1 dapat dijadikan ke bentuk statistik (mean, std, min, max, dst)
```

```python
dfdrop2 = df['price'].dropna() # mencari nilai df['price'] yang di-dropna
dfdrop2
```

```python
dfdrop2.describe() # melihat apakah variabel dfdrop2 dapat dijadikan ke bentuk statistik (mean, std, min, max, dst)
```

```python
dfdrop3 = df['perc_of_unit_sold'].dropna() # mencari nilai df['perc_of_unit_sold'] yang di-dropna agar hasil perhitungannya dapat dimasukkan ke nilai Fillna-nya
dfdrop3
```

```python
df3x = pd.to_numeric(dfdrop3, errors = 'coerce') # mengubah variabel dfdrop3 yang berformat 'object' menjadi 'float64' agar dapat dicari statistiknya
df3x
```

```python
df3x.describe() # melihat apakah variabel df3x dapat dijadikan ke bentuk statistik (mean, std, min, max, dst)
```

```python
dffill3 = df['perc_of_unit_sold'].fillna(df3x.mean()) # mencari nilai df['perc_of_unit_sold'] yang di-fillna dengan memasukkan nilai mean dari dropna-nya
dffill3
```

```python
df3y = pd.to_numeric(dffill3, errors = 'coerce') # mengubah variabel df3y yang berformat 'object' menjadi 'float64' agar dapat dicari statistiknya
df3y
```

```python
df3y.describe() # melihat apakah variabel df3y dapat dijadikan ke bentuk statistik (mean, std, min, max, dst)
```

```python
dfdrop4 = df['rating'].dropna() # mencari nilai df['rating'] yang di-dropna agar hasil perhitungannya dapat dimasukkan ke nilai Fillna-nya
dfdrop4
```

```python
df4x = pd.to_numeric(dfdrop4, errors = 'coerce') # mengubah variabel dfdrop4 yang berformat 'object' menjadi 'float64' agar dapat dicari statistiknya
df4x
```

```python
df4x.describe() # melihat apakah variabel df4x dapat dijadikan ke bentuk statistik (mean, std, min, max, dst)
```

```python
dffill4 = df['rating'].fillna(df4x.mean()) # mencari nilai df['rating'] yang di-fillna dengan memasukkan nilai mean dari dropna-nya
dffill4
```

```python
df4y = pd.to_numeric(dffill4, errors = 'coerce') # mengubah variabel df4y yang berformat 'object' menjadi 'float64' agar dapat dicari statistiknya
df4y
```

```python
df4y.describe() # melihat apakah variabel df4y dapat dijadikan ke bentuk statistik (mean, std, min, max, dst)
```

## Analisis dan Visualisasi Data

### Nilai-nilai yang telah di-Dropna atau Fillna dimasukkan sebagai variabel baru pada dfbaru

```python
dfbaru = df 
dfbaru['Original_price'] = dfdrop1
dfbaru['price'] = dfdrop2
dfbaru['perc_of_unit_sold'] = df3y
dfbaru['rating'] = df4y
```

```python
dfbaru.corr() # Mencari korelasi dari masing-masing variabel yang telah di-dropna atau di-fillna
```

```python
sns.heatmap(dfbaru.corr()) # Mencari korelasi dari masing-masing variabel yang telah di-dropna atau di-fillna dalam bentuk heatmap
```

```python
sns.heatmap(df.corr(), cmap='coolwarm', annot=True) # Mencari korelasi dari masing-masing variabel yang telah di-dropna atau di-fillna dalam bentuk heatmap yang mengandung nilai korelasi
```

### Dari hasil pemetaan dengan heatmap, didapatkan bahwa variabel 'Original_price' dan 'price' memiliki korelasi yang besar pada nilai 0.71. Hal tersebut masih masuk akal karena keuntungan dihitung dari selisih antara harga produk ketika dibeli (Original_price) dengan harga produk ketika dijual (price).

### Variabel perc-of-unit_sold tidak berpengaruh banyak terhadap Original_price dan rating sehingga korelasinya dapat diabaikan. Suatu catatan adalah variabel perc_of_unit_sold memiliki korelasi negatif dengan variabel 'price'. Dengan demikian, persentase barang yang terjual akan semakin besar jika harga jual produk semakin murah, dan sebaliknya.

### Variabel rating ternyata tidak memiliki korelasi yang signifikan. Namun, perlu diperhatikan bahwa korelasinya terhadap 'Original_price' dan 'price' bertanda negatif. Jadi, customer memberikan rating yang lebih tinggi untuk produk dengan harga jual dan harga beli yang lebih murah.

```python
sns.distplot(dfbaru['Original_price']) # Melihat pemetaan distribusi untuk setiap variabel
plt.show()
```

### Dari pemetaan distribusi 'Original_price', variabel banyak terkonsentrasi pada 'Original_price' yang rendah. Dengan demikian, penjualan produk dengan Original_price yang nilainya kecil/murah menjadi lebih besar dibandingkan ketika 'Original_price' lebih tinggi

```python
sns.distplot(dfbaru['price'])
plt.show()
```

### Bentuk pemetaan distribusi 'price' mirip dengan bentuk pemetaan pada Original_price. Namun, jika dilihat dari penyebaran horizontalnya, produk masih terjual pada price sebesar 1500. Hanya saja, ada perbedaan yang sangat besar untuk Original_price (nilai max sekitar 30000-an) dengan price (nilai max di antara 3000 - 3500). E-Commerce dapat dikatakan menderita kerugian apabil persebaran price tidak lebih besar dari penyebaran Original_price. Untuk meningkatkan keuntungan, E-Commerce dapat mengurangi produk dengan harga yang sangat tinggi dan lebih memusatkan penjualan produk yang Original_price-nya lebih murah.

```python
sns.distplot(dfbaru['perc_of_unit_sold'])
plt.show()
```

### Bentuk pemetaan perc_of_unit_sold mendekati distribusi normal dengan peak/puncak pada nilai perc_of_unit_sold di kisaran 60 %. Maka, persentase barang yang terjual hampir tersebar merata. Secara ideal, bentuk pemetaan adalah dengan nilai peak pada perc_of_unit_sold sebesar-besarnya (mendekati 100). Penjualan produk E-Commerce secara keseluruhan dapat dikatakan memuaskan, namun masih ada produk dengan persentase penjualan yang relatif rendah.

```python
sns.distplot(dfbaru['rating'])
plt.show()
```

### Secara keseluruhan, bentuk pemetaan dari variabel rating adalah memuaskan dengan peak pada kisaran 85 - 90. Dari pemetaan distribusi rating, ternyata tingkat kepuasan konsumen adalah cukup puad atau bagus.

```python
sns.pairplot(dfbaru, palette = 'coolwarm')
```

```python
P = sns.PairGrid(dfbaru)

P.map_diag(plt.hist)
P.map_upper(plt.scatter)
P.map_lower(plt.scatter)
```

### Dari hasil pemetaan pairplot dan scatterplot, terlihat bahwa produk terkonsentrasi pada nilai price dan Original_price yang rendah. Hal itu sesuai dengan minat konsumen yang lebih menyukai membeli produk dengan harga yang semakin kecil/murah. 

### Dari pemetaan perc_of_unit_sold, ternyata penyebarannya merata pada rating yang cukup tinggi. Namun, ada sebagian kecil customer yang memberi rating rendah.

### Kesimpulan Akhir: E-Commerce mendapat rating yang memuaskan dari customer, tetapi keuntungan yang didapat masih dapat dimaksimalkan lagi. E-Commerce sebaiknya memperbanyak membeli produk yang harga belinya murah dan mengurangi pembelian produk yang harga belinya tinggi.
